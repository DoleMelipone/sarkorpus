TF-IDF by year (2002-2025) and source (Libération, Le Monde, Le Figaro), for nouns :

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from adjustText import adjust_text

path1 = os.path.join('tf-idf', 'tf-idf_par_an_&_source_1&2')
path2 = os.path.join('tf-idf_title', 'tf-idf_par_an_&_source_1&2')
path3 = os.path.join('tf-idf_all', 'tf-idf_par_an_&_source_1&2')
```

```python
def graph_tfidf(path, trunc=False):
    if trunc:
        years = list(range(2015, 2026))
    else:
        years = list(range(2002, 2026))
    eq = {'figaro': 'Le Figaro', 'monde': 'Le Monde', 'liberation': 'Libération'}
    res = {'liberation': [], 'monde': [], 'figaro': []}
    for dir in years:
        for key in res:
            mot = ''
            val = 0
            if trunc or not (dir == 2024 and key == 'liberation'):
                fpath = os.path.join(path, str(dir), key+'_tfidf.tsv')
                df = pd.read_csv(fpath, sep='\t', header=None)
                for w in df.itertuples():
                    if w[1].split('_')[0] not in ['sarkozy', 'nicolas-sarkozy', 'affaire']:
                        mot = w[1].split('_')[0]
                        val = w[2]
                        break
            res[key].append((mot,val))

    plt.style.use("seaborn-v0_8-whitegrid")

    fig, axes = plt.subplots(
        3, 1,
        figsize=(14,10),
        sharex=True,
        sharey=True
    )

    colors = {
        'figaro': '#1f77b4',
        'monde': '#2ca02c',
        'liberation': '#d62728'
    }

    for ax, journal in zip(axes, res):

        data = res[journal]
        vals = [v[1] for v in data]
        mots = [v[0] for v in data]

        ax.plot(
            years,
            vals,
            marker='o',
            linewidth=2,
            markersize=4,
            color=colors[journal]
        )

        texts = []

        for x, y, mot in zip(years, vals, mots):
            texts.append(
                ax.text(x, y, mot, fontsize=12, color=colors[journal])
            )

        adjust_text(
            texts,
            ax=ax,
            arrowprops=dict(arrowstyle="-", lw=0.5, color="gray")
        )

        ax.set_title(eq[journal])
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

    axes[-1].set_xlabel("Year")

    fig.supylabel("TF-IDF score")
    #fig.suptitle("Word with max TF-IDF by newspaper and by year", fontsize=14)

    plt.tight_layout()
    plt.show()
```
